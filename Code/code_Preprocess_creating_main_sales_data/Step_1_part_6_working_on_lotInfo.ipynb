### --------- Convert "RAW_dictionary_bbrInfoBox" column to dictionary object                

```python
def decode_json(x):
    try:
        if isinstance(x, str) and x.strip() != "" and not pd.isna(x):
            return ast.literal_eval(x)
            # Alternatively, you can use json.loads as follows:
            # return json.loads(x.replace("'", '"'))
        else:
            return x
    except (ValueError, SyntaxError) as e:
        print(f"Error decoding JSON: {e}")
        return None

# Apply the decoding function to the specified column
data_lotInfo['RAW_dictionary_lotInfo'] = data_lotInfo['RAW_dictionary_lotInfo'].apply(decode_json)
```

# ---- Create swimming pool dummy variable based on values in Notes column

```python
data_lotInfo['lotInfo_swimmingPool_dummy'] = 0
```

```python
swimming_keyWord = ['swimmingpool','svømmepool','svømmebassin']
def swimming_dummy(x):
    swimming = False
    try:
        if isinstance(x, list):
            for list_item in x:
                list_string = list_item.lower()

                if any(keyword in list_string for keyword in swimming_keyWord):
                    swimming = True
                    break
                
            if swimming:
                return 1
            else:
                return 0

        else:
            return 0
    except (ValueError, SyntaxError) as e:
        print(f"Error decoding JSON: {e}")
        return None

# Apply the decoding function to the specified column
data_lotInfo['lotInfo_swimmingPool_dummy'] = data_lotInfo['notes'].apply(swimming_dummy)
```

#### ------- create variable: greenhouse Dummy

```python
data_lotInfo['lotInfo_greehouse_dummy'] = 0  
```

```python
def greenhouse_dummy(x):
    swimming = False
    try:
        if isinstance(x, list):
            for list_item in x:
                list_string = list_item.lower()

                if 'drivhus' in list_string:
                    swimming = True
                    break
                
            if swimming:
                return 1
            else:
                return 0

        else:
            return 0
    except (ValueError, SyntaxError) as e:
        print(f"Error decoding JSON: {e}")
        return None

# Apply the decoding function to the specified column
data_lotInfo['lotInfo_greehouse_dummy'] = data_lotInfo['notes'].apply(greenhouse_dummy)
```
