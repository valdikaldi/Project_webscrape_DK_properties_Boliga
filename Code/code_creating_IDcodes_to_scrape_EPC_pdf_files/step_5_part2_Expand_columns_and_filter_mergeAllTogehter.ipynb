```python
import pandas as pd

# Define chunk size
chunk_size = 10000 # 1031595/10000 == 104 chunks
# Adjust this value based on your memory constraints



# Function to merge chunks
def merge_chunks(main_chunk, col1_chunk, col2_chunk, col3_chunk, col4_chunk, col5_chunk, col6_chunk):
    
    print('--------- Mergin column_1')
    # Merge column_1 with main_file.csv based on 'gui_code'
    merged_chunk = pd.merge(main_chunk, col1_chunk, on='guid_code', how='left')
    
    print('--------- Mergin column_2')
    # Merge column_2 with the merged chunk based on 'gui_code'
    merged_chunk = pd.merge(merged_chunk, col2_chunk, on='guid_code', how='left')
    
    print('--------- Mergin column_3')
    # Merge column_3 with the merged chunk based on 'gui_code'
    merged_chunk = pd.merge(merged_chunk, col3_chunk, on='guid_code', how='left')
    
    print('--------- Mergin column_4')
    # Merge column_4 with the merged chunk based on 'gui_code'
    merged_chunk = pd.merge(merged_chunk, col4_chunk, on='guid_code', how='left')
    
    print('--------- Mergin column_5')
    # Merge column_5 with the merged chunk based on 'gui_code'
    merged_chunk = pd.merge(merged_chunk, col5_chunk, on='guid_code', how='left')
    
    print('--------- Mergin column_6')
    # Merge column_6 with the merged chunk based on 'gui_code'
    merged_chunk = pd.merge(merged_chunk, col6_chunk, on='guid_code', how='left')
    
    return merged_chunk



main_columns = [
#     'Unnamed: 0',                   
    'guid_code',                    
    'property_esrCode',             
    'municipalityCode',             
    'addressString',                
    'streetname',                   
    'address_number',               
    'address_floor_number',         
    'addressFloor',                 
    'address_door_number',          
    'Zip_code_name',                
    'Zip_code',                     
    'is_on_sale',                   
    'propertyType_Name',            
    'propertyType_code',            
    'isApartment',                  
    'is_cooperative_property',      
#     'char_bbrInfoBox',              
#     'char_buildingInfo',            
#     'char_unitInfo',                
#     'char_lotInfo',                 
#     'char_cadastralInfos',          
#     'char_additionalBuldingInfos',  
#     'char_floorInfos',              
#     'char_additionalFloorInfos',    
    'Price_Transaction_sales',      
    'Price_Public_evaluations',     
    'Price_current_listing',        
    'Price_previous_listing',       
    'latitude',                     
    'longitude',                    
    'bfenr',                        
    'ejendomsværdiskat',            
    'grundskyld'                   
]

# ------------------------- READ: Main file - part_1_Boliga_nameChanged_columnsDeleted.csv
path_main=r'D:\Thesis\Properties\Denmark\RE_due_scraping_properties\Boliga_dk\Step_5_Expand_propertyChar_columns_and_filter\part_1_Boliga_nameChanged_columnsDeleted.csv'
main_chunks = pd.read_csv(path_main, encoding='utf-8',chunksize=chunk_size, usecols=main_columns)

# ------------------------- READ: Column_1_char_bbrInfoBox_EXPANDED.csv
path_1=r'D:\Thesis\Properties\Denmark\RE_due_scraping_properties\Boliga_dk\Step_5_Expand_propertyChar_columns_and_filter\Ready\Column_1_char_bbrInfoBox_EXPANDED_Ready_noindex.csv'
col1_chunks = pd.read_csv(path_1, encoding='utf-8', chunksize=chunk_size)

# ------------------------- READ: Column_2_char_buildingInfo_EXPANDED.csv
path_2=r'D:\Thesis\Properties\Denmark\RE_due_scraping_properties\Boliga_dk\Step_5_Expand_propertyChar_columns_and_filter\Ready\Column_2_char_buildingInfo_EXPANDED_Ready_noindex.csv'
col2_chunks = pd.read_csv(path_2, encoding='utf-8', chunksize=chunk_size)

# ------------------------- READ: Column_3_char_unitInfo_EXPANDED.csv
path_3=r'D:\Thesis\Properties\Denmark\RE_due_scraping_properties\Boliga_dk\Step_5_Expand_propertyChar_columns_and_filter\Ready\Column_3_char_unitInfo_EXPANDED_Ready_noindex.csv'
col3_chunks = pd.read_csv(path_3, encoding='utf-8', chunksize=chunk_size)

# ------------------------- READ: Column_4_char_lotInfo_EXPANDED.csv
path_4=r'D:\Thesis\Properties\Denmark\RE_due_scraping_properties\Boliga_dk\Step_5_Expand_propertyChar_columns_and_filter\Ready\Column_4_char_lotInfo_EXPANDED_Ready_noindex.csv'
col4_chunks = pd.read_csv(path_4, encoding='utf-8', chunksize=chunk_size)

# ------------------------- READ: Column_5_char_cadastralInfos_EXPANDED.csv
path_5=r'D:\Thesis\Properties\Denmark\RE_due_scraping_properties\Boliga_dk\Step_5_Expand_propertyChar_columns_and_filter\Ready\Column_5_char_cadastralInfos_EXPANDED_Ready_noindex.csv'
col5_chunks = pd.read_csv(path_5, encoding='utf-8', chunksize=chunk_size)

# ------------------------- READ: Column_6_char_additionalBuldingInfos_EXPANDED.csv
path_6=r'D:\Thesis\Properties\Denmark\RE_due_scraping_properties\Boliga_dk\Step_5_Expand_propertyChar_columns_and_filter\Ready\Column_6_char_additionalBuldingInfos_EXPANDED_Ready_noindex.csv'
col6_chunks = pd.read_csv(path_6, encoding='utf-8', chunksize=chunk_size)




# counter for print
counter = 0 

# Iterate through chunks of main_file.csv
for main_chunk in main_chunks:
    counter +=1
    print(f'--------------------------- Chunk number {counter} out of 104 chunks (10.000 rows per chunk)')
    
    result_chunks = []

    # Iterate through chunks of column_1.csv and column_2.csv to find a match
    for col1_chunk, col2_chunk, col3_chunk,col4_chunk,col5_chunk,col6_chunk in zip(col1_chunks, col2_chunks, col3_chunks, col4_chunks, col5_chunks, col6_chunks):
        # Merge the chunks
        merged_chunk = merge_chunks(main_chunk, col1_chunk, col2_chunk, col3_chunk,col4_chunk,col5_chunk,col6_chunk)
        result_chunks.append(merged_chunk)
        break  # Break after the first match, assuming each 'gui_code' is unique

    # Concatenate the result chunks into the final result
    final_result = pd.concat(result_chunks, ignore_index=True)

    # Save the final result to main_file.csv
    path_final = r'D:\Thesis\Properties\Denmark\RE_due_scraping_properties\Boliga_dk\Step_5_Expand_propertyChar_columns_and_filter\Final\Boliga_Final_merged_char_variables.csv'
    final_result.to_csv(path_final, mode='a', encoding='utf-8', header=not os.path.exists(path_final), index=False)
```
