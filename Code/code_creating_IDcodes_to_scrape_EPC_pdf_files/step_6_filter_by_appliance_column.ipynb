# number of NORMAL transaction sales from the 981.469 properties 

```python
path = r'D:\Thesis\Properties\Denmark\RE_due_scraping_properties\Boliga_dk\Step_6_filter_by_appliance\Boliga_Final_filter_by_appliance.csv'
chunk_size = 10000

final_data_chunks = pd.read_csv(path, encoding='utf-8', low_memory=False, chunksize=chunk_size)

counter = 0 
count_normal_sales = 0
for chunk in final_data_chunks:
    counter+=1
    print(f'------------- chunk {counter} out of 99 (981469/10000) ')
    
    for index,row in chunk.iterrows():
        sales_string = row['Price_Transaction_sales']
        
        dictionary = ast.literal_eval(sales_string)

        # count normal sales
        for obj in dictionary:
            if obj['handoverName'] == 'Alm. frit salg':
                count_normal_sales +=1
                
print(f'---Total number of normal sales from 981.469 residential properties is: {count_normal_sales}')

---Total number of normal sales from 981.469 residential properties is: 2.295.344 
```

# Number of NORMAL transation sales from 981.469 properties per year 

```python
path = r'D:\Thesis\Properties\Denmark\RE_due_scraping_properties\Boliga_dk\Step_6_filter_by_appliance\Boliga_Final_filter_by_appliance.csv'
chunk_size = 10000

final_data_chunks = pd.read_csv(path, encoding='utf-8', low_memory=False, chunksize=chunk_size)

year_list = []

counter = 0 
count_normal_sales = 0
for chunk in final_data_chunks:
    counter+=1
    print(f'------------- chunk {counter} out of 99 (981469/10000) ')
    
    for index,row in chunk.iterrows():
        sales_string = row['Price_Transaction_sales']
        
        dictionary = ast.literal_eval(sales_string)

        # count normal sales
        for obj in dictionary:
            date_string = obj['recalculationDate'] 
            if '2006' in date_string : year_list.append('2006') 
            if '2007' in date_string : year_list.append('2007') 
            if '2008' in date_string : year_list.append('2008') 
            if '2009' in date_string : year_list.append('2009') 
            if '2010' in date_string : year_list.append('2010') 
            if '2011' in date_string : year_list.append('2011') 
            if '2012' in date_string : year_list.append('2012') 
            if '2013' in date_string : year_list.append('2013') 
            if '2014' in date_string : year_list.append('2014') 
            if '2015' in date_string : year_list.append('2015') 
            if '2016' in date_string : year_list.append('2016') 
            if '2017' in date_string : year_list.append('2017') 
            if '2018' in date_string : year_list.append('2018') 
            if '2019' in date_string : year_list.append('2019') 
            if '2020' in date_string : year_list.append('2020') 
            if '2021' in date_string : year_list.append('2021') 
            if '2022' in date_string : year_list.append('2022') 
            if '2023' in date_string : year_list.append('2023') 
                
from collections import Counter

# Count unique values using Counter
counted_values = Counter(year_list)
counted_values


###################################################### result #############################################################
result: 
         '2006': 123.707
         '2021': 119.431
         '2007': 113.604
         '2020': 108.036
         '2017': 102.606
         '2018': 99.655
         '2019': 97.511
         '2016': 92.656
         '2015': 92.567
         '2010': 92.039
         '2008': 83.579
         '2022': 78.140
         '2014': 75.499
         '2009': 69.562
         '2011': 67.748
         '2013': 67.431
         '2012': 66.202
         '2023': 54.115
```
